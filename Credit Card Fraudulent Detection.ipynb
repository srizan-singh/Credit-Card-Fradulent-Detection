# Using Under-Sampling Method

Under-Sampling

Build a sample dataset containing similar distribution of normal(Legit) transactions and Fraudulent(Fraud)Transactions

Number of Fraudulent Transactions --> 492

```python
Legit_sample=Legit.sample(n=492)
```

concatenating two Dataframes

axis=0--->ROWS

axis=1--->COLOUMNS

```python
new_dataset=pd.concat([Legit_sample,Fraud],axis=-0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

# Splitting the Dataframe into Target and features

```python
x=new_dataset.drop(columns='Class',axis=1)
y=new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

# Split the Dataset into Training Data and Testing Data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

test_size=0.2 --->20% data will go for Testing and 80% data will go under Training

```python
print(x.shape,x_test.shape,x_train.shape)
```

# Model Training (training the model)

Logistic Model

Training the Logistic training Model with Traiing data

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

# Model Evaluation

Accuracy Score

# Accuracy on training data

```python
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

# Accuracy on testing data

```python
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
